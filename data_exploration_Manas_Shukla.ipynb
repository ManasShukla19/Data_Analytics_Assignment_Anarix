```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
data = pd.read_csv('Global-Superstore - Global-Superstore.csv')
```

```python
# Display the first 5 rows of the dataset
data.head()
```

```python
# Get an overview of the dataset
data.info()
```

```python
# Check for missing values
data.isnull().sum()
```

```python
# Drop the 'Postal Code' column due to missing values
data.drop("Postal Code", axis=1, inplace=True)
data.isnull().sum()
```

```python
# Check data types and basic statistics
data.dtypes
data.describe()
```

```python
# Total Sales by Market
graph = data.groupby('Market')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
graph.plot(kind='bar')
plt.title('Total Sales by Market')
plt.xlabel('Market')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## This bar plot shows the total sales for each market. APAC has the highest sales, indicating significant market demand.

```python
# Total Sales by Region
graph = data.groupby('Region')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
graph.plot(kind='bar')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## This bar plot illustrates total sales by region. The 'Central' region has the highest sales, highlighting a key area for business.

```python
# Shipping Cost vs Region
graph = data.groupby('Region')['Shipping Cost'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
graph.plot(kind='bar')
plt.title('Shipping Cost vs Region')
plt.xlabel('Region')
plt.ylabel('Shipping Cost')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## This bar plot shows total shipping costs by region. The 'Central' region incurs the highest shipping costs, potentially due to higher sales volumes.

```python
# Order Priority Analysis
priority = data['Order Priority'].value_counts()
priority.plot(kind='bar')
plt.title('Order Priority Distribution')
plt.xlabel('Priority Level')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## This bar plot shows the distribution of order priorities. 'Medium' priority orders are the most common.

```python
# Order Priority Pie Chart
labels = data['Order Priority'].unique()
sizes = data['Order Priority'].value_counts()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Order Priority Distribution (Pie Chart)')
plt.show()
```

## The pie chart provides a visual distribution of order priorities, with 'Medium' priority being the most prevalent.

```python
# Discount and Profit Analysis
plt.scatter(data['Discount'], data['Profit'])
plt.title('Discount vs. Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

## The scatter plot shows the relationship between discount and profit. A negative correlation indicates higher discounts reduce profit margins.

```python
# Product Analysis by Sub-Category
products = data.groupby('Sub-Category')['Sales'].sum().sort_values()
sns.barplot(x=products, y=products.index)
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sales')
plt.ylabel('Sub-Category')
plt.show()
```

## This bar plot displays total sales by product sub-category. The 'Phones' sub-category has the highest sales, indicating high demand.

```python
# Shipping and Delivery Analysis
# Calculate the frequency of each ship mode
ship_mode_counts = data['Ship Mode'].value_counts()
# Calculate the average shipping cost for each ship mode
average_shipping_cost = data.groupby('Ship Mode')['Shipping Cost'].mean()

# Plotting the frequency of ship modes
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, plot 1
ship_mode_counts.plot(kind='bar', color='lightblue')
plt.title('Frequency of Different Ship Modes')
plt.xlabel('Ship Mode')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

# Plotting the average shipping cost by ship mode
plt.subplot(1, 2, 2)  # 1 row, 2 columns, plot 2
average_shipping_cost.plot(kind='bar', color='lightgreen')
plt.title('Average Shipping Cost by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Average Shipping Cost')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## The first bar plot shows the frequency of different shipping modes, with 'Standard Class' being the most used.
## The second bar plot shows average shipping costs, with 'Same Day' shipping being the most expensive.

```python
# Customer Segmentation
segments = data['Segment'].value_counts()
plt.pie(segments, labels=segments.index, autopct='%1.1f%%', startangle=140)
plt.title('Customer Segment Distribution')
plt.show()
```

## The pie chart shows the distribution of customer segments. 'Consumer' is the largest segment, followed by 'Corporate' and 'Home Office'.

```python
# Customer Segmentation: Order Counts and Total Sales
order_counts = data['Segment'].value_counts()
total_sales = data.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, plot 1
order_counts.plot(kind='bar', color='lightblue')
plt.title('Count of Orders by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)  # 1 row, 2 columns, plot 2
total_sales.plot(kind='bar', color='lightgreen')
plt.title('Total Sales by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## The first bar plot shows the number of orders per customer segment, with 'Consumer' segment having the highest number of orders.
## The second bar plot shows total sales per segment, with 'Consumer' also generating the highest sales.

```python
# Correlation Analysis
correlation_matrix = data[['Sales', 'Profit', 'Discount', 'Quantity']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Sales, Profit, Discount, and Quantity')
plt.show()
```

## The heatmap reveals relationships between key variables. Sales and profit have a strong positive correlation, while discount and profit have a negative correlation.

```python
# Time Series Analysis
data['Order Date'] = pd.to_datetime(data['Order Date'])
monthly_sales = data.resample('M', on='Order Date')['Sales'].sum()

plt.figure(figsize=(14, 7))
plt.plot(monthly_sales)
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

## The time series plot shows monthly sales trends over time, highlighting seasonality and growth patterns.
